# RestAPIs

## no parameters: Kanye API
- Simple restAPI 
- Call using requests library
- extract json resonse

```python
import requests
import pandas as pd
```

```python
api_url = "https://api.kanye.rest/"
response = requests.get(api_url)
response = response.json()['quote']
```

```python
response
```

## with parameters: Cocktail API

```python
url = 'https://www.thecocktaildb.com/api/json/v1/1/search.php?s=margarita'
cocktail = requests.get(url).json()
cocktail
```

## API key authentification: Holiday API

```python
from requests.auth import HTTPBasicAuth
auth = HTTPBasicAuth('api_key', '')
```

```python
holidays = requests.get('https://calendarific.com/api/v2/holidays?&country=US&year=2019', auth=auth)
```

```python
holidays.json()
```

## API with O Authentication: The Noun Project 


```python
import requests
from requests_oauthlib import OAuth1

auth = OAuth1("", "")
endpoint = "https://api.thenounproject.com/v2/icon/1"

response = requests.get(endpoint, auth=auth)
print(response.content)
```
